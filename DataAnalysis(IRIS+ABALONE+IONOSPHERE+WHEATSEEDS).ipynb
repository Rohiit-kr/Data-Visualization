```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

IRIS FLOWERS DATASET

```python
iris= pd.read_csv('/content/drive/MyDrive/VAC/IRIS.csv')
iris.head()
```

```python
iris.describe()
```

```python
iris.isnull().sum()
```

```python
iris.groupby('species').agg(['mean','max','min'])
```

```python
iris.sort_values(by='species', ascending=False)
```

```python
iris[iris['petal_length'] == iris['petal_length'].max()]
```

```python
iris[iris['petal_length'] == iris['petal_length'].min()]
```

```python
sns.countplot(x='species', data=iris, hue='sepal_length')
plt.title('Species Distribution by Sepal Length')
```

```python
corr= iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr()
sns.heatmap(corr, cmap='viridis')
```
